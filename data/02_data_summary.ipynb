```python
import os

def summarize_data_in_folders(root_dir):
    # Iterate through all subdirectories in the root directory
    for subdir, _, files in os.walk(root_dir):
        print(f"Processing folder: {subdir}")
        
        # Iterate through the files in the current subdirectory
        for filename in files:
            # Check if the file is a .jsonl file
            if filename.endswith(".jsonl"):
                file_path = os.path.join(subdir, filename)
                try:
                    # Open the file and count the lines
                    with open(file_path, 'r') as file:
                        lines = file.readlines()
                        print(f"File: {filename} - Number of lines: {len(lines)}")
                except Exception as e:
                    print(f"Error reading {filename}: {e}")
        print("")

# Specify the root directory
root_directory = "/work/van-speech-nlp/jindaznb/jslpnb/mllm_experiments/slam-llm/data/"

# Call the function with the root directory
summarize_data_in_folders(root_directory)
```
