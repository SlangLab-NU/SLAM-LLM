```python
import os
import json
from datasets import Dataset

# Function to update the dataset and write it into a JSONL file
def process_dataset_and_save_jsonl(dataset, dataset_name, output_dir):
    """Process the dataset and save it as a JSONL file."""
    
    # Initialize a list to store the processed data
    data = []
    
    # Iterate through all examples in the dataset
    for example in dataset:
        file = example['file']
        audio = example['audio']
        text = example['text'].lower()  # Convert text to lowercase
        speaker_id = example['speaker_id']
        chapter_id = example['chapter_id']
        example_id = example['id']
        file_name = example['id'].split('-')[-1]
        
        # Construct the file path dynamically based on speaker_id and chapter_id
        new_file_path = os.path.join(
            f"/work/van-speech-nlp/librispeech/LibriSpeech/{dataset_name}/{speaker_id}/{chapter_id}/{example_id}.flac"
        )
        
        # Check i f the file exists
        if os.path.exists(new_file_path):
            print(f"File exists: {new_file_path}")
            example['file'] = new_file_path
            
            # Prepare the JSONL entry
            jsonl_entry = {
                "key": example_id,
                "source": new_file_path,
                "target": text
            }
            data.append(jsonl_entry)
        else:
            print(f"File does not exist: {new_file_path}")
    
    # Define output path and create the directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    jsonl_output_path = os.path.join(output_dir, f"{dataset_name}.jsonl")
    
    # Write data to JSONL file
    with open(jsonl_output_path, 'w') as f:
        for entry in data:
            f.write(json.dumps(entry) + '\n')
    
    print(f"Processed {len(data)} entries. JSONL file saved at: {jsonl_output_path}")


output_directory = "librispeech-100_test_path"

# Process both datasets and save them as JSONL
process_dataset_and_save_jsonl(librispeech_train_100, "train-clean-100", output_directory)
process_dataset_and_save_jsonl(librispeech_val_360, "dev-clean", output_directory)
```
