```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity

def plot_similarity_matrix(speech_embeddings, text_embeddings, title):
    # Calculate cosine similarity
    similarity_matrix = cosine_similarity(speech_embeddings, text_embeddings)
    
    # Create heatmap
    plt.figure(figsize=(10, 8))
    plt.imshow(similarity_matrix, cmap='viridis', aspect='auto')
    plt.colorbar()
    plt.title(title)
    plt.xlabel('Text Tokens')
    plt.ylabel('Speech Tokens')
    plt.tight_layout()
    plt.show()

# Example usage (you would replace these with actual embeddings)
speech_embeddings = np.random.rand(20, 128)  # 20 speech tokens, 128-dim embeddings
text_embeddings = np.random.rand(15, 128)    # 15 text tokens, 128-dim embeddings

plot_similarity_matrix(speech_embeddings, text_embeddings, "Before LoRA")

# Simulate LoRA effect (you would actually apply LoRA and get new embeddings)
speech_embeddings_lora = np.random.rand(20, 128)
text_embeddings_lora = np.random.rand(15, 128)

plot_similarity_matrix(speech_embeddings_lora, text_embeddings_lora, "After LoRA")
```
